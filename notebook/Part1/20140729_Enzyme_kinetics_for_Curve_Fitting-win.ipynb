## Enzyme kinetics

Enzyme kinetics is the study of the chemical reactions that are catalysed by enzymes. In enzyme kinetics, the **reaction rate** is measured and the effects of varying the conditions of the reaction are investigated. Studying an enzyme's kinetics in this way can reveal the catalytic mechanism of this enzyme, its role in metabolism, how its activity is controlled, and how a drug or an ahonist might inhibit the enzyme.

## General principles

The reaction catalysed by an enzyme uses exactly the same reactants and produces exactly the same products as the uncatalysed reaction. Like other catalysts, enzymes do not alter the position of equilibrium between substrates and products. However, unlike uncatalysed chemical reactions, enzyme-catalysed reactions display saturation kinetics.

## Single-substrate reactions

### Michaelis-Menten Kinetics

As enzyme-catalysed reactions are saturable, their rate of catalysis does not show a linear response to increasing substrate. If the initial rate of the reaction is measured over a range of substrate concentrations (denoted as [S]), the reaction rate (v) increases as [S] increases, as shown below. However, as [S] gets higher, the enzyme becomes saturated with substrate and the rate reaches V<sub>max</sub>, the enzyme's maximum rate.

<img src = 'Images\MiMen.png' />
<img src = 'Images\EKequ.png' />

**The Michaelis-Menten kinetic model of a single-substrate reaction** is shown above. The Michaelis-Menten equation describes how the initial reaction rate v<sub>0</sub> depends on the position of the substrate-binding equilibrium and the rate constant k<sub>2</sub>.<br><br>

$$ v = \dfrac{V_{max}[S]}{K_{M}+[S]} $$ <br>

<center>with the constants</center>

$$ K_{M} = \dfrac{k_{2} + k_{-1}}{k_{1}} \simeq K_{D} $$ <br>
$$ V_{max} = k_{cat}[E]_{total} $$ <br><br>

```python
from xlrd import open_workbook
import numpy as np
from scipy.optimize import  leastsq

# Import data from Excel file
mmdata = open_workbook('data\MichaelisMenten.xlsx')
mmsheet = mmdata.sheet_by_index(0)

con, rr = [], []

for row_index in range(mmsheet.nrows):
    con.append(mmsheet.cell(row_index, 0).value)
    rr.append(mmsheet.cell(row_index, 1).value)
    
plot(con, rr, 'o')
```

```python
con = np.array(con)  # applied substrate concentrations
rr = np.array(rr) # measured reaction rates
n_rr = (rr/0.6) / (3.6e-2) # normalized reaction rates
```

```python
def Michaelis_Menten(x, Km, Vmax):
    """
    Michaelis Menten Equation
    """
    return (Vmax * x) / (Km + x)
```

```python
def residuals(p, y, x):
    """
    Deviations of data from fitted Michaelis Menten curve
    """
    Km, Vmax = p
    err = y - Michaelis_Menten(x, Km, Vmax)
    return err
```

```python
def peval(x, p):
    """
    Evaluated value at x with current parameters.
    """
    Km, Vmax = p
    return Michaelis_Menten(x, Km, Vmax)
```

```python
# Make up some data for fitting

x = np.linspace(0,80,50)
Km, Vmax = 11, 1.4
y_true = Michaelis_Menten(x, Km, Vmax)
```

```python
# Initial guess for parameters

p0 = [8, 1.2]
```

```python
# Fit equation usint least squares optimization

plsq = leastsq(residuals, p0, args=(n_rr, con))
```

```python
# Plot
plot(con, n_rr, 'o', x, peval(x,plsq[0]), x, y_true)

# Add legend on left curves
legend(['Fit', 'Measured', 'Initial'], loc='upper left', )
```

# Measurement of Enzyme Kinetics using Isothermal Titration Calorimetry

The primary experimental observable is heat-flow measureed in μW (μJs<sup>-1</sup>). this heat-flow is directly proportional to the reaction rate.
However, calculation of the reaction rate requires knowledge of the enthalpy change of the reaction (ΔH<sub>r</sub>) and the volume (V) of the reaction chamber.

The enthalpy change is calculated by measurements of the total-heat, which is consumed or released and the total number of moles which are converted from substrate to product:


$$\Delta H_{r} = \dfrac{1}{n_{sub}}\int ^{t=\infty }_{t=0}\dfrac{dQ}{dt}dt $$

Where n<sub>sub</sub> is the number of moles of substrate which are converted to products and dQ/dt is the heat-flow. Thus the reaction rate (𝑣) can be calculated:

$$ v = \dfrac{1}{V \Delta H_{r}} \dfrac{dQ}{dt} $$

The substrate concentration ([sub]) as a function of time can be vcalculated using the integrated heat as a function of time:

$$ [sub](t) = [sub]_{t=1} - \dfrac{\int^{t}_{t=0} \dfrac{dQ}{dt}dt}{\Delta H_{r}V} $$

Using last two Eqs. it is possible to calculate both the reaction rate and substrate concentration as a function of time and finding the kinetic parameters k<sub>cat</sub> and K<sub>M</sub> by fitting to the Michaelis-Menten equation.

```python
from xlrd import open_workbook
import numpy as np

# Import data from Excel file

data = open_workbook('data\itc_gox.xlsx')
sheet = data.sheet_by_index(0)

t, Q = [], []

for row_index in range(sheet.nrows):
    t.append(sheet.cell(row_index, 0).value)
    Q.append(sheet.cell(row_index, 1).value)
    
```

```python
# Change list to numpy array
t = np.array(t)
Q = np.array(Q)
plot(t,Q)

# Baseline fitting and convert Unit (Calorie to Jule)
Q = ((Q-Q[0])*1E-6)*4.184
```

```python
from scipy.integrate import cumtrapz

# Total number of substrate mole and cell volume
nsub = 1E-3
V = 1.4

# Calculate Total Entalpy Changes
delta_Hr = 1/nsub * sum(cumtrapz(Q, dx = 2))

# Calculate reaction rate
reaction_rate = 1/(1.4*delta_Hr)*Q

# Calculate substrate concentration
def cal_con(time, nsub, Q, delta_Hr, V):
    return nsub - (sum(cumtrapz(Q[:time], dx = 2))/(delta_Hr*V))

con = []

for i in t:
    con.append(cal_con(i, nsub, Q, delta_Hr, V))
```

```python
plot(con, reaction_rate)
xlabel('Glucose concentration($M$)', fontsize = 15)
ylabel(r'Reaction rate($JM^{-1}min^{-1}$)', fontsize = 15)
xlim(0.0, 0.0010)
ylim(0e-9, 7e-9)
```
