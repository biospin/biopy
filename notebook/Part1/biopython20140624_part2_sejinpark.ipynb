### Reverse Complement sequence를 만들어봅시다. 

```python
from IPython.display import Image
Image(url='http://freepages.genealogy.rootsweb.ancestry.com/~langolier/IMG/DNA/comp_strands.gif')
```

```python
buf = []
for i in seq:
    if i == 'A':
        buf.append('T')
    elif i == 'T':
        buf.append('A')
    elif i == 'G':
        buf.append('C')
    elif i == 'C':
        buf.append('G')
    else :
        buf.append('N')
buf.reverse()
```

```python
rc_dna = "".join(buf)
```
